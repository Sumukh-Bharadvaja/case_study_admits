* The main focus is to create a demographics breakdown of one of the program based on ethnicity or gender or both
* This business case involves concentrating on groups in order to promote education as well as to provide special aid and care with minority.

Importing the Necessary Libraries

```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import textdistance

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r"Sample_Data.csv")
#Show the data
data
```

As the data neededin order to show the demographics breakdown needes to be cleaned and saved separately we need to perform the following steps as previously done to the main data

```python
data_clean=data[['Program_Desc','IPEDS_Ethnicities','IPEDS_Hispanic','Sex']]
```

```python
data_clean
```

```python
data_clean=data_clean.dropna()

```

```python
unique_branch = ['Accounting MS','Business Administration - MBA','Real Estate Development - MS','Business - PHD','Finance - MS','Management - MS','Technology Management - MS']

comparision_string=None 

def compare_with_array(column_element):
    global comparision_string
    return textdistance.jaccard(column_element,comparision_string)
```

```python
Program_name=pd.DataFrame()

for unique_values in unique_branch:
    comparision_string = unique_values
    Program_name[unique_values] = data_clean['Program_Desc'].apply(compare_with_array)

Program_name['Final']=Program_name[['Accounting MS','Business Administration - MBA','Real Estate Development - MS','Business - PHD','Finance - MS','Management - MS','Technology Management - MS']].idxmax(axis=1)
Program_name
        
# data_1['Accounting MS']=data_1['Program_Desc'].apply(compare_with_array)
data_clean['Program_Desc_Final']=Program_name['Final']

```

```python
data_clean = data_clean.drop('Program_Desc', axis=1)
```

```python
data_eth=data_clean[['IPEDS_Ethnicities','Program_Desc_Final']]
data_his=data_clean[['IPEDS_Hispanic','Program_Desc_Final']]
data_sex=data_clean[['Sex','Program_Desc_Final']]
```

Now that we have saved into three different dataframes , since the MBA admission are more, let us concentrate more on the MBA data only. Filtering can be done as shown below.

## Final Ethnicity breakdown 

```python
data_eth_final=data_eth[data_eth['Program_Desc_Final']=='Business Administration - MBA']
data_eth_final.to_csv("Final_Ethenicity.csv")
```

## Final hispanic/Non-Hispanic breakdown 

```python
data_his_final=data_his[data_his['Program_Desc_Final']=='Business Administration - MBA']
data_his_final.to_csv("His_final.csv")
```

## Final Gender breakdown 

```python
data_sex_final=data_sex[data_sex['Program_Desc_Final']=='Business Administration - MBA']
data_sex_final.to_csv("Sex_final.csv")
```
